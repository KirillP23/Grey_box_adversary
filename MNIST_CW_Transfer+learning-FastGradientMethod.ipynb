# Let's pick 10 legit examples to target:

````python
nb_adv_per_sample = str(nb_classes - 1)
print('Crafting ' + str(source_samples) + ' * ' + nb_adv_per_sample + ' adversarial examples')
###########################################################################

assert source_samples == nb_classes
idxs = [np.where(np.argmax(Y_test, axis=1) == i)[0][0] for i in range(nb_classes)]
# [3, 2, 1, 18, 4, 8, 11, 0, 61, 7]
# Pick the first appearances of the 9 numbers in X_test
for i,idx in enumerate(idxs):
    plt.subplot(2,5,i+1)
    plt.imshow(X_test[idx,:,:,0],cmap='gray')
    plt.axis('off')
````

# Create adv samples using the Carlini-L2 attack:

````python
# Form a 10 x 10 array of the inputs that you seek to target:
adv_inputs = np.array([[instance] * nb_classes for instance in X_test[idxs]],dtype=np.float32)
adv_inputs = adv_inputs.reshape((source_samples * nb_classes, img_rows, img_cols, 1))


one_hot = np.zeros((nb_classes, nb_classes))
one_hot[np.arange(nb_classes), np.arange(nb_classes)] = 1



adv_ys = np.array([one_hot] * source_samples,dtype=np.float32).reshape((source_samples * nb_classes, nb_classes))
# yname = "y_target"
# Instantiate a CW attack object
#model1 = KerasModelWrapper(model)
#(Vishal) model passed into FastGradientMethod must be of type Model (as defined) so we pass our keras model to the KerasModelWrapper
model_wrap=KerasModelWrapper(model)
cw = FastGradientMethod(model_wrap, back='tf', sess=sess)
#(Vishal) Original binary_search_steps and max_iterations used
cw_params = {'y_target': adv_ys,}

#{'binary_search_steps': 15,
#             'y_target': adv_ys,
#             'max_iterations': 50000,
#             'learning_rate': 1.5,
#             'batch_size': source_samples * nb_classes if
#             targeted else source_samples,
#             'initial_const': 1,}

adv = cw.generate_np(adv_inputs, **cw_params)
````

* In cell 12, Matrix norm dist b/w adv and input image is very different. For FastGradientMethod, it is constant while in CarliniWagnerL2, it fluctuates. <br /> 
* In cell 17,Matrix norm dist b/w adv and input image is very similar between FastGradientMethod and CarliniWagnerL2. <br />
* In cell 18, we see the deviation between is not less than 0.1 between any adv and input image unlike in CarliniWagnerL2.  <br />
* In cell 21, While predicting for the adv_nt and adv_inputs and see how many match, we find only 8 match as compared to 12 in CarliniWagnerL2.  <br />
* The plots from cell 26,27 are identical in both cases as expected (data is the same).<br />
* In general, the models tend to predict slightly worse as compared to the one in CarliniWagnerL2 but are still very much comparable. The only exception is when training on the adversarial features where the FastGradientMethod does noticably worse.<br />
* In cell 30, this can be seen in the Histogram of MNIST trained model predictions for FMNIST dataset.(the distributions are very different)<br />
* In cell 35,36, we notice the adv images generated by FastGradientMethod, is more noisy as compared to their counterparts in CarliniWagnerL2, while they both dont match at all (in cell 37,38)<br />
* We also see while the clean images distribution match (in cell 40), the adversarial images distribution dont match. In general,FastGradientMethod seems to be adding greater noise to the images which may explain why the models do not predict as well
